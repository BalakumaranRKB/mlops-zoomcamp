```python
import pandas as pd
```

```python
df_yellow_JAN2023 = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2024-01.parquet')
```

```python
df_yellow_FEB2023 = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2024-02.parquet')
```

```python
df_yellow_JAN2023
```

 there are 19 columns in the Yellow taxi January dataset

```python
df_yellow_JAN2023.columns
```

```python
df_yellow_JAN2023['duration'] = (df_yellow_JAN2023.tpep_dropoff_datetime - df_yellow_JAN2023.tpep_pickup_datetime)/pd.Timedelta(minutes=1)
```

```python
df_yellow_JAN2023
```

```python
df_yellow_JAN2023['duration'].std()
```

```python
df_new_yellow_JAN2023 = df_yellow_JAN2023.loc[(df_yellow_JAN2023['duration']>=1) & (df_yellow_JAN2023['duration']<=60)]
```

```python
(df_new_yellow_JAN2023.shape[0]/df_yellow_JAN2023.shape[0])*100
```

```python
df= df_new_yellow_JAN2023[['PULocationID', 'DOLocationID']]
```

```python
df
```

```python
df.info()
```

```python
df = df.astype(str)
```

```python
df.info()
```

```python
input_data = df.to_dict('records')
```

```python
input_data
```

```python
from sklearn.feature_extraction import DictVectorizer


# Step 2: Fit a dictionary vectorizer
dv = DictVectorizer(sparse=True)  # Set sparse=False to get a dense numpy array

# Step 3: Get a feature matrix from it
X = dv.fit_transform(input_data)

# To see the dimensionality (number of columns)
print(f"Dimensionality of the one-hot encoded matrix: {X.shape[1]}")

# To see the feature names (column names)
print(dv.get_feature_names_out())
```

```python
X 
```

```python
y = df_new_yellow_JAN2023['duration']
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X, y)
```

```python
y_pred = reg.predict(X)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
rmse = mean_squared_error(y, y_pred, squared=False)
print(f"RMSE: {rmse}")
```
